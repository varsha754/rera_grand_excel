```python
# prompt: my data frame have some values like 6/2/2024 and some like 30-12-2022, i want all values like 30-12-2022

import pandas as pd
from datetime import datetime

# ... (Your existing code)

def change_date(x):
    if pd.isna(x):
        return x

    if isinstance(x, datetime):
        return x.strftime("%d-%m-%Y")
    elif isinstance(x, str):
        try:
            datetime.strptime(x, "%d-%m-%Y")
            return x
        except ValueError:
            try:
                datetime.strptime(x, "%d/%m/%Y")
                return datetime.strptime(x, "%d/%m/%Y").strftime("%d-%m-%Y")
            except ValueError:
                return x
    else:
        return x


# Apply the updated change_date function to all relevant columns
latest['Proposed Date of Completion'] = latest['Proposed Date of Completion'].apply(change_date)
latest['Commencement bottom left date'] = latest['Commencement bottom left date'].apply(change_date)
latest['Commencement Date'] = latest['Commencement Date'].apply(change_date)
latest['Rera_commmencing_ending_with'] = latest['Rera_commmencing_ending_with'].apply(change_date)
latest['Last_Modified_Date'] = latest['Last_Modified_Date'].apply(change_date)
latest['revised date of completion'] = latest['revised date of completion'].apply(change_date)
```
